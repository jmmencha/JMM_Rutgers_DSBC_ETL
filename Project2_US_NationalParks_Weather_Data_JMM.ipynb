```python
# Update "Carlsbad Caverns National Park" latitude & longitudinal coordinate combination to that of the nearest major city,
# "Whites City, New Mexico 88220". Coordinates to be used by the "Open Weather Map" API to locate weather data
# for "Carlsbad Caverns National Park".

# Carlsbad Caverns National Park:
# DMS: 32° 10′ 31″ N, 104° 26′ 38″ W
# Decimal: 32.175278, -104.443889

# Whites City, New Mexico 88220 
# DMS: 32° 10′ 32″ N, 104° 22′ 36″ W
# Decimal: 32.175556, -104.376667

# Set "Carlsbad Caverns National Park" coordinates equal to that of "Whites City, New Mexico 88220" coordinates:
dw_updated_df.iloc[[2], [1]] = 32.175556
dw_updated_df.iloc[[2], [2]] = -104.376667
dw_updated_df
```

```python
# Update "National Park of American Samoa" latitude & longitudinal coordinate combination to that of the nearest major city,
# "Pago Pago, Eastern District, American Samoa". Coordinates to be used by the "Open Weather Map" API to locate weather
# data for "National Park of American Samoa".

# National Park of American Samoa:
# DMS: 14° 15′ 30″ S, 170° 41′ 0″ W
# Decimal: -14.258333, -170.683333

# Pago Pago, Eastern District, American Samoa:
# DMS: 14° 16′ 46″ S, 170° 42′ 2″ W
# Decimal: -14.279444, -170.700556

# Set "National Park of American Samoa" coordinates equal to that of "Pago Pago, Eastern District, American Samoa" coordinates:
dw_updated_df.iloc[[1], [1]] = -14.279444
dw_updated_df.iloc[[1], [2]] = -170.700556
dw_updated_df
```

```python
# List for holding latitude & longitudinal coordinates, and city names:
lat_lngs = []
cities = []

# Note: The "dw_updated_df" DataFrame columns "coordinates_latitude" & "coordinates_longitude" are the (center point)
# latitude & longitudinal coordinate combination for a given U.S. National Park.

# Create a "Numpy Array" (data type: numpy.ndarray) named "lats", from the "dw_updated_df" DataFrame
# column "coordinates_latitude":
lats = dw_updated_df['National Park Latitude Coordinate'].to_numpy()
print(f"\lats:\n{lats}\ntype:{type(lats)}")

# Create a "Numpy Array" (data type: numpy.ndarray) named "lngs", from the "dw_updated_df" DataFrame
# column "coordinates_longitude":
lngs = dw_updated_df['National Park Longitude Coordinate'].to_numpy()
print(f"\lngs:\n{lngs}\ntype:{type(lngs)}")
```

```python
# Create a "zip" object named "lat_lngs", from Numpy Arrays "lats" & "lngs":
# ("lat_lngs" to be passed to the "Open Weather Map" API)
lat_lngs = zip(lats, lngs)

# Identify the nearest city to each U.S. National Park's latitude & longitudinal coordinate combination:
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
# Append each city to the list of cities:
    cities.append(city)

# Print the city count to confirm sufficient count:
len(cities)
```

```python
# Print out to verify data:
print(lat_lngs)
print(lats)
print(lngs)
type(lat_lngs)
```

```python
# Base "Open Weather Map" URL:
url = "http://api.openweathermap.org/data/2.5/weather?"
```

```python
# Initialize response value lists:
city_with_data = []
cloudiness = []
country = []
date = []
humidity = []
lat = []
lng = []
max_temp = []
wind_speed = []
cities_to_remove = []

print("------------------------------------------------------------")
print("Begin Data Retrieval:")
print("------------------------------------------------------------")

# Loop through the list of cities and request "response value" data for each city in the list:
for city in range(len(cities)):
    query_url = url + "appid=" + weather_api_key + "&units=" + "Imperial" + "&q=" + cities[city]
    response = requests.get(query_url)
    response_json = response.json()
    
    try:
        # Weather data collection for each city in the list:
        city_with_data.append(response_json["name"])
        cloudiness.append(response_json["clouds"]["all"])
        country.append(response_json["sys"]["country"])
        date.append(response_json["dt"])
        humidity.append(response_json["main"]["humidity"])
        lat.append(response_json["coord"]["lat"])
        lng.append(response_json["coord"]["lon"])
        max_temp.append(response_json["main"]["temp_max"])
        wind_speed.append(response_json["wind"]["speed"])
        
        # Print out a log for each city in the list:
        print(f"Processing Record {city} | {cities[city]}")
        
    except (KeyError, IndexError):
        print(f"Processing Record {city} | {cities[city]}")
        print("The listed city is not significant enough in size, to be part of the Open Weather Map API's dataset(s).")
        print("Therefor, this city will be skipped, as no weather data is currently availalbe for this 'remote' location.")
        
print("------------------------------------------------------------")
print("Completed Data Retrieval.")
print("------------------------------------------------------------")
```

```python
# Print out the number of data successfully collected from the Open Weather Map API for each response value list:
print(f"City: {len(city_with_data)}")
print(f"Country: {len(country)}")
print(f"Latitude: {len(lat)}")
print(f"Longitude: {len(lng)}")
print(f"Date: {len(date)}")
print(f"Wind Speed: {len(wind_speed)}")
print(f"Cloudiness: {len(cloudiness)}")
print(f"Humidity: {len(humidity)}")
print(f"Max Temp: {len(max_temp)}")
```

```python
# Create an "Open Weather Map - U.S. National Parks Weather Data" DataFrame named "owm_npwd_df",
# from the collected response data:
owm_npwd_df = pd.DataFrame({
    "City": city_with_data,
    "Country": country,
    "Latitude": lat,
    "Longitude": lng,    
    "Date": date,
    "Wind Speed": wind_speed,
    "Cloudiness": cloudiness,
    "Humidity": humidity,    
    "Max Temp": max_temp,    
})
owm_npwd_df
```

```python
# Add each value of the "dw_updated_df" DataFrame's "National Park Name" column to the list,
# "dw_updated_df__us_nationalpark_title_list":
dw_updated_df__us_nationalpark_title_list = dw_updated_df['National Park Name'].tolist()
dw_updated_df__us_nationalpark_title_list
```

```python
# Error Correction:
# The Open Weather Map API cannot find weather data near American Samoa National Park, as it is too remote a location
# (out of range) for "citipy" to retrieve the "nearest [major] city" latitude & longitudinal coordinate combination.

# Remove "National Park of American Samoa" from the list, "dw_updated_df__us_nationalpark_title_list":
dw_updated_df__us_nationalpark_title_list.remove('American Samoa')
dw_updated_df__us_nationalpark_title_list
```

```python
# Insert "dw_updated_df__us_nationalpark_title_list", as a new column into the "owm_npwd_df" DataFram,
# and name it "National Park Name":
owm_npwd_df.insert(0, "National Park Name", dw_updated_df__us_nationalpark_title_list, True)
owm_npwd_df
```

```python
# Begin the process of converting the "owm_npwd_df" DataFrame's "Date" column values from unix format to a readable format.
# Add each value of the "owm_npwd_df" DataFrame's "Date" column to the list, "owm_npwd_unix_date_list":
owm_npwd_unix_date_list = owm_npwd_df['Date'].tolist()
owm_npwd_unix_date_list
```

```python
# Using "datetime" from "datetime" module, convert date format from unix to readable for each item in list,
# "owm_npwd_unix_date_list", and store it in list "owm_npwd_readable_date_list":
owm_npwd_readable_date_list = []

# Iterate through the list of dates, "owm_npwd_unix_date_list":
for time_stamp in owm_npwd_unix_date_list:
    
    # Convert timestamp back to datetime object:
    date_time_object = datetime.fromtimestamp(time_stamp)
    
    # Append each date, as a string, to the list " owm_npwd_readable_date_list":      
    owm_npwd_readable_date_list.append(str(date_time_object))
    
# Print list:
owm_npwd_readable_date_list
```

```python
# Replace the "owm_npwd_df" DataFrame's "Date" column (unix format), with list, "owm_npwd_readable_date_list" (readable):
owm_npwd_df['Date'] = owm_npwd_readable_date_list
owm_npwd_df
```

```python
# Export "owm_npwd_df" DataFrame to CSV file named "OpenWeatherMap_US_NationalParks_WeatherData.csv":
output_data_file = "./output_data/OpenWeatherMap_US_NationalParks_WeatherData.csv"
owm_npwd_df.to_csv(output_data_file, index = False, header = True)
```
