```python
# Create a list of the "gsmnptdt_df" DataFrame columns:
cols = gsmnptdt_df.columns.tolist()

# Print the "cols" list (CURRENT order of data columns within the "gsmnptdt_df" DataFrame):
cols
```

```python
# Reorder the "gsmnptdt_df" DataFrame columns to match that of the other National Park Service (NPS) DataFrames:
cols_reordered = ['OBJECTID', 'TRAILNAME', 'TRLALTNAME', 'TRLLABEL', 'TRLFEATTYPE', 'TRAILSTATUS', 'TRLSURFACE', 'TRAILTYPE',
         'TRAILCLASS', 'TRLUSE', 'UNITCODE', 'UNITNAME', 'EDITDATE', 'MAPMETHOD', 'MAPSOURCE', 'SOURCEDATE', 'XYERROR',
         'GEOMETRYID', 'DISTRIBUTE', 'RESTRICTION', 'ISEXTANT', 'GROUPCODE', 'REGIONCODE', 'CREATEDATE', 'SOURCESCALE',
         'FMSSID', 'ASSETID', 'NOTES', 'FEATUREID', 'EDITBY', 'CREATEBY', 'FACILITYTYPE', 'ASSETCODE', 'COUNTY', 'LANDFORM',
         'MANAGEMENTZONE', 'WATERSHED', 'LON', 'LAT', 'COORD_SYSTEM', 'DATUM', 'UTM_ZONE', 'X_COORD', 'Y_COORD', 'STATE',
         'QUADNAME', 'HERROR', 'VERROR', 'PARKDISTRICT', 'FMSS_LOC', 'FMSS_ASSET', 'LOC_NAME', 'MTFCC', 'FRA', 'ACCESS',
         'AGENCY_DAT', 'ACCESS_STA', 'ADMIN_ORG', 'CONG_DIST', 'DESIGN_USE', 'HIST_SIGNF', 'JURIDICT', 'LAND_PLAN',
         'MANGED_US', 'MTR_PROHIBITED', 'NAT_TR_DES', 'PR_TR_MNTR', 'PROHIB_USE', 'ROW', 'ROAD_SYS', 'SHARED_SYS',
         'SPC_MGT_AR', 'GRSM_STATUS', 'RuleID', 'Shape__Length']

# Print the "cols_reordered" list (NEW order of data columns within the "gsmnptdt_df" DataFrame):
cols_reordered
```

```python
# Re-create the "gsmnptdt_df" DataFrame as "gsmnptdt_cols_reordered_df", where this new DataFrame has its data columns
# in order according to the "cols_reordered" list:
gsmnptdt_cols_reordered_df = gsmnptdt_df[cols_reordered]

# Print the "gsmnptdt_cols_reordered_df" DataFrame:
gsmnptdt_cols_reordered_df
```

```python
# Drop all unwanted extraneous columns from the "gsmnptdt_cols_reordered_df" DataFrame & rename it as
# "gsmnptdt_cols_dropped_df":
gsmnptdt_cols_dropped_df = gsmnptdt_cols_reordered_df.drop(['GEOMETRYID', 'DISTRIBUTE', 'RESTRICTION', 'ISEXTANT',
        'GROUPCODE', 'REGIONCODE', 'CREATEDATE', 'SOURCESCALE', 'FMSSID', 'ASSETID', 'NOTES', 'FEATUREID', 'EDITBY',
        'CREATEBY', 'FACILITYTYPE', 'ASSETCODE', 'COUNTY', 'LANDFORM', 'MANAGEMENTZONE', 'WATERSHED', 'LON', 'LAT',
        'COORD_SYSTEM', 'DATUM', 'UTM_ZONE', 'X_COORD', 'Y_COORD', 'STATE', 'QUADNAME', 'HERROR', 'VERROR',
        'PARKDISTRICT', 'FMSS_LOC', 'FMSS_ASSET', 'LOC_NAME', 'MTFCC', 'FRA', 'ACCESS', 'AGENCY_DAT', 'ACCESS_STA',
        'ADMIN_ORG', 'CONG_DIST', 'DESIGN_USE', 'HIST_SIGNF', 'JURIDICT', 'LAND_PLAN', 'MANGED_US', 'MTR_PROHIBITED',
        'NAT_TR_DES', 'PR_TR_MNTR', 'PROHIB_USE', 'ROW', 'ROAD_SYS', 'SHARED_SYS', 'SPC_MGT_AR', 'GRSM_STATUS',
        'RuleID', 'Shape__Length'], axis=1)

# Print the "gsmnptdt_cols_dropped_df" DataFrame:
# Note: Columns "INPLACES" & "GlobalID" do not exist in the "Great_Smoky_Mountains_National_Park_3D_Trails.csv" file,
# and thus also do not exist in the "gsmnptdt_df", "gsmnptdt_cols_reordered_df", or the 
# "gsmnptdt_cols_dropped_df" DataFrames.
gsmnptdt_cols_dropped_df
```

```python
# Save "gsmnptdt_cols_dropped_df" as "gsmnptdt_cols_added_df":
gsmnptdt_cols_added_df = pd.DataFrame(gsmnptdt_cols_dropped_df)

# Insert two "NaN" (empty) columns of an arbitrary length into the new DataFrame, "gsmnptdt_cols_added_df":
gsmnptdt_cols_added_df["INPLACES"] = np.nan
gsmnptdt_cols_added_df["GlobalID"] = np.nan
gsmnptdt_cols_added_df
```

```python
# Save "gsmnptdt_cols_added_df" as "gsmnptdt_cols_renamed_df":
gsmnptdt_cols_renamed_df = pd.DataFrame(gsmnptdt_cols_added_df)

# Rename columns in the "gsmnptdt_cols_renamed_df" DataFrame so they match EXACTLY with the column names
# in the "nps_all_df" DataFrame:
gsmnptdt_cols_renamed_df.rename(index = str, columns = {"TRAILNAME": "TRLNAME", "TRAILSTATUS": "TRLSTATUS",
                                         "TRAILTYPE": "TRLTYPE", "TRAILCLASS": "TRLCLASS"}, inplace = True)
gsmnptdt_cols_renamed_df
```

```python
# Save "gsmnptdt_cols_renamed_df" as "gsmnptdt_cleaned_df":
gsmnptdt_cleaned_df = pd.DataFrame(gsmnptdt_cols_renamed_df)
gsmnptdt_cleaned_df
```

```python
# Merge all National Park Service (NPS) DataFrames into one DataFrame, "nps_all_df":
nps_all_df = pd.concat([anmt_df, bcnpt_df, clnpt_df, crnpt_df, cbnmt_df, dnmt_df, gcnrat_df, gsnhst_df,
                        gsmnptdt_cleaned_df, hnmt_df, nbnmt_df, rbnmt_df, tcnmt_df, znpt_df], axis = 0)
nps_all_df
```

```python
# Reset the index of the "nps_all_df" DataFrame so that the index increments by one for each row of data:
nps_all_df.reset_index(drop = True, inplace = True)

# Print the "nps_all_df" DataFrame:
#pd.set_option('display.max_columns', None)
#pd.set_option('display.max_rows', None)
nps_all_df
```

```python
# Rename "nps_all_df" DataFrame columns:
# "UNITNAME" ---> "NationalParkName"

nps_all_df.columns = ['OBJECTID', 'TRLNAME', 'TRLALTNAME', 'TRLLABEL', 'TRLFEATTYPE', 'TRLSTATUS',
                      'TRLSURFACE', 'TRLTYPE', 'TRLCLASS', 'TRLUSE', 'UNITCODE', 'NationalParkName', 'EDITDATE',
                      'MAPMETHOD', 'MAPSOURCE', 'SOURCEDATE', 'XYERROR', 'INPLACES', 'GlobalID']
nps_all_df
```

```python
# Reorder "nps_all_df" DataFrame columns:
nps_all_df = nps_all_df[['NationalParkName', 'OBJECTID', 'TRLNAME', 'TRLALTNAME', 'TRLLABEL', 'TRLFEATTYPE',
                         'TRLSTATUS', 'TRLSURFACE', 'TRLTYPE', 'TRLCLASS', 'TRLUSE', 'UNITCODE',
                         'EDITDATE', 'MAPMETHOD', 'MAPSOURCE', 'SOURCEDATE', 'XYERROR', 'INPLACES', 'GlobalID']]
nps_all_df
```

```python
# Export the "nps_all_df" DataFrame to a CSV file named "NationalParkService(NPS)_US_NationalParks_TrailsData.csv":
output_data_file = "./output_data/NationalParkService(NPS)_US_NationalParks_TrailsData.csv"
nps_all_df.to_csv(output_data_file, index = False, header = True)
```
