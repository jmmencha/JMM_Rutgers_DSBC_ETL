```python
# Perform an "outer merge" using Pandas, (also referred to as a “full outer join” or just “outer join”).
# Joining "irma_1979_df" and "irma_1980_to_1989_df" via "outer merge":
irma_1979_through_1989_df = pd.merge(irma_1979_df, irma_1980_to_1989_df, on = 'Park Name', how = 'outer')

print("There are {} unique values of 'Park Name' in the dataframes."
      .format(pd.concat([irma_1979_df['Park Name'], irma_1980_to_1989_df['Park Name']]).unique().shape[0]))

print("Outer merge result has ('rows', 'columns') = {}."
      .format(irma_1979_through_1989_df.shape))

print("There are {} rows with no missing values."
      .format((irma_1979_through_1989_df.apply(lambda x: x.isnull().sum(), axis = 1) == 0).sum()))
```
