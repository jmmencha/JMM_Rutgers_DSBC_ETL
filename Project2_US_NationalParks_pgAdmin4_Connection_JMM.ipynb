```python
# Send "Trails Data", "Visitors Data", "Summary Data", and "Weather Data" DataFrames to pgAdmin 4
# (Management tools for PostgreSQL) DataBase (DB) "Prj2_US_NtlPks_pgAdmin4_PostgreSQL_db":

# Connect to the DB:
engine = create_engine(f'postgresql://{user}:{password}@localhost:5432/Prj2_US_NtlPks_pgAdmin4_PostgreSQL_db')
```

```python
# Access table names within the pgAdmin 4 DB, "Prj2_US_NtlPks_pgAdmin4_PostgreSQL_db":
engine.table_names()
```

```python
# National Park Service (NPS):
# Send this Jupyter Notebook's (Trails Data) "nps_all_df" DataFrame to the pgAdmin 4 table, "nps_all", within the DB:
nps_all_df.to_sql(name = 'nps_all',
                  con = engine,
                  if_exists = 'append',
                  index = False)
```

```python
# Open Weather Map:
# Send this Jupyter Notebook's (Weather Data) "owm_npwd_df" DataFrame to the pgAdmin 4 table, "owm_npwd", within the DB:
owm_npwd_df.to_sql(name = 'owm_npwd',
                   con = engine,
                   if_exists = 'append',
                   index = False)
```
